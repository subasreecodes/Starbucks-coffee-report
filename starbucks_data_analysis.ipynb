# Starbucks Data Analysis

```python
#import the required modules
import pandas as pd
import numpy as np
```

```python
#read the file
starbucks_df = pd.read_csv('starbucks.csv')
```

```python
#view first few records
starbucks_df.head(10)
```

```python
#check for any null values
starbucks_df.isnull().any()
```

```python
#column with null values
starbucks_df['Caffeine (mg)']
```

```python
#record with NaN value
starbucks_df[starbucks_df['Caffeine (mg)'].isnull()]
```

```python
#to fill the NaN value, target the specific category and analyze to fill
starbucks_df[starbucks_df['Beverage_category']=='Shaken Iced Beverages']
```

```python
#creating a copy before making changes, final clean dataset
starbucks_data = starbucks_df.copy()
```

```python
#forward fill method 
starbucks_data['Caffeine (mg)'].ffill
```

```python
#after ffill method
starbucks_data[starbucks_data['Caffeine (mg)'].isnull()]
```

```python
starbucks_df[starbucks_df['Beverage_category']=='Shaken Iced Beverages']
```

```python
#information about data
starbucks_data.info()
```

```python
#statistical overview
starbucks_data.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Set the style for seaborn
sns.set(style="whitegrid")

# Create a bar plot showing the average calories for each beverage
plt.figure(figsize=(10, 6))
sns.barplot(x='Beverage', y='Calories', data=starbucks_data, ci=None)

# Rotate the x-axis labels for better readability if there are many beverages
plt.xticks(rotation=90, ha="right")

# Set plot title and labels
plt.title('Average Calories per Beverage at Starbucks')
plt.xlabel('Beverage')
plt.ylabel('Calories')

# display the plot
plt.tight_layout()
plt.show()
```
